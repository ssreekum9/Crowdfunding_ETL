### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
kk = crowdfunding_info_df.columns
print(kk)
```

```python

crowdfunding_info_df[["category","subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n=1, expand=True)
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.
import pandas as pd

def projecttwo_lists(df, column_name):
    return set(df[column_name])

categories = projecttwo_lists(crowdfunding_info_df,'category')
subcategories = projecttwo_lists(crowdfunding_info_df,'subcategory')

print(categories)
print(subcategories)
```

```python

print(len(categories))
print(len(subcategories))
```

```python

category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
category_ids = range(1, 10) 
subcategory_ids = range(1, 25)

category_prefix = 'categ'
categoryy_ids = [category_prefix + str(x) for x in category_ids]

subprefix = 'subcat'
subids = [subprefix + str(x) for x in subcategory_ids]

print(categoryy_ids)
print(subids)
```

```python

category_df = pd.DataFrame(list(zip(categoryy_ids, categories)), \
                          columns=['category_id', 'category'])

subcategory_df = pd.DataFrame(list(zip(subids, subcategories)), \
                          columns=['subcategory_id', 'subcategory']) 



```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.

campaign_df = campaign_df.rename(columns={'blurb': 'description', 'deadline': 'end_date', 'launched_at': 'launch_date'})
campaign_df.head()
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df['pledged'] = campaign_df['pledged'].astype(float)
campaign_df['goal'] = campaign_df['goal'].astype(float)

campaign_df.head(13)
                                                           
```

```python
# Check the datatypes
q = campaign_df.dtypes
print(q)
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt

campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'])
campaign_df['launch_date'] = pd.to_datetime(campaign_df['launch_date'])

print(campaign_df.head())
```

```python
# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
campaign_merged_df = pd.merge(campaign_df, category_df, on='category', how='left')
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, on='subcategory', how='left')

campaign_merged_df.tail(10)
```

```python
# Drop unwanted columns
campaign_cleaned = campaign_merged_df.drop(['staff_pick', 'spotlight'], axis=1)

campaign_cleaned.head()

```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

### Option 2: Use regex to create the contacts DataFrame.

```python
contact_info_df_copy = contact_info_df.copy()
contact_info_df_copy.head()
```

```python
# Extract the four-digit contact ID number.
import re 

contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_info'].apply(lambda i: re.search(r'"contact_id": (\d{4})', i).group(1) if re.search(r'"contact_id": (\d{4})', i) else None) #Code snippet taken from: https://www.w3schools.com/python/python_regex.asp
contact_info_df_copy.head(9)
```

```python
# Check the datatypes.
contact_info_df_copy.dtypes
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df_copy['contact_id'] = contact_info_df_copy['contact_id'].astype('int64')
```

```python
contact_info_df_copy['name'] = contact_info_df_copy['contact_info'].apply(lambda i: re.search(r'"name": "([^"]+)"', i).group(1) if re.search(r'"name": "([^"]+)"', i) else None)# Extract the name of the contact and add it to a new column.

```

```python

contact_info_df_copy['email'] = contact_info_df_copy['contact_info'].apply(lambda i: re.search(r'"name": "([^"]+)"', i).group(1) if re.search(r'"name": "([^"]+)"', i) else None)
```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contact_info_df_copy_2 = contact_info_df_copy[['contact_id', 'name', 'email']].copy()
contact_info_df_copy_2
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column.  

contact_info_df_copy_2[['first_name', 'last_name']] = contact_info_df_copy_2['name'].str.split(' ', n=1, expand=True)
contact_info_df_copy_2 = contact_info_df_copy_2.drop('name', axis=1) 
print(contact_info_df_copy_2.head())

```

```python
# Reorder the columns
col_new = ['contact_id', 'first_name', 'last_name', 'email']

contact_info_df_copy_2 = contact_info_df_copy_2.reindex(columns=col_new)

contact_info_df_copy_2.head()
```

```python
# Check the datatypes one more time before exporting as CSV file.
gj = contact_info_df_copy_2.dtypes
print(gj)
```

```python
# Export the DataFrame as a CSV file. 
contact_info_df_copy_2.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
